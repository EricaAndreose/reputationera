```python
import pandas as pd

df = pd.read_csv('bibliographic_data_KLI.csv')


filtered_df = df[df['date'].notnull() & df['date'].astype(str).str.strip().str.replace('*', '').str.replace('-', '').str.isnumeric()]

# Filter out rows with invalid 'date' values (including '*')
#filtered_df['date'] = filtered_df['date'].str.replace('*', '')

# Convert 'date' column to integer type
filtered_df['date'] = filtered_df['date'].astype(int)

# Filter rows based on the date column
filtered_df = filtered_df[filtered_df['date'].between(2004, 2014)]
filtered_df.reset_index(drop=True, inplace=True)

filtered_df

```
