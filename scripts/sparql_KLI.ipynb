```python
import sparql_dataframe
import pandas as pd
# query wikidata 
endpoint = "https://query.wikidata.org/bigdata/namespace/wdq/sparql"

q = """ 
select ?label where {wd:Q34661 wdt:P6379 ?o .
                ?o rdfs:label ?label
                    FILTER(lang(?label) = 'en')}
"""

df = sparql_dataframe.get(endpoint, q, post=True)
df
```

```python
endpoint = "https://query.wikidata.org/bigdata/namespace/wdq/sparql"

q = """ 
SELECT  ?award ?awardLabel ?date 
WHERE
{
  wd:Q34661 p:P166 ?awardstatement .
  ?awardstatement ps:P166 ?award .
  ?awardstatement pq:P585 ?date .
  SERVICE wikibase:label { bd:serviceParam wikibase:language "en,fr" . }
}
"""

df = sparql_dataframe.get(endpoint, q, post=True)
df
```

```python
endpoint = "https://query.wikidata.org/bigdata/namespace/wdq/sparql"

q = """ 
SELECT DISTINCT ?label ?date ?directorname WHERE {
  ?s wdt:P31 wd:Q11424;
     wdt:P577 ?date;
     wdt:P57 ?director;
     rdfs:label ?label.
  ?director rdfs:label ?directorname.
  FILTER(CONTAINS(LCASE(?label), "klimt"@en))
  FILTER(LANGMATCHES(LANG(?directorname), "en"))
}
"""

df = sparql_dataframe.get(endpoint, q, post=True)
df
```
